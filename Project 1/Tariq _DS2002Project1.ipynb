```python
#i.2
#Convert from CSV to JSON
csv_file = 'pokemonGO.csv' #original CSV file
json_file = 'pogo.json' #new JSON file

pogo_json = {}
with open(csv_file, 'r') as pogocsv:
    reader = csv.DictReader(pogocsv)
    for row in reader:
        # Use one of the CSV column names as a key, pokemon name
        key = row['Name']
        pogo_json[key] = row 
with open(json_file,'w') as pogocsv:   #writing the CSV file onto a JSON file
    pogocsv.write(json.dumps(pogo_json, indent=2))
    
pogo_json
```

```python
try:    #checking for errors
  print(pogo_json)
except:
  print("An error occurred")
```

```python
#creating a new empty database to convert from CSV to SQL
from sqlite3 import Error


def create_connection(pogo_db):
    """ create a database connection to a SQLite database """
    conn = None
    try:
        conn = sqlite3.connect(pogo_db)   #opens connection to an SQL database
        print(sqlite3.version)  #check for errors
    except Error as e: 
        print(e)
    finally:
        if conn:
            conn.close()


if __name__ == '__main__':
    create_connection(r"/Users/sharaftariq/Downloads/pogosql.db") 
```

```python
#i.2
#converting to SQL
con = sqlite3.connect('pogosql.db')
# creating connection with newly made empty database
cur = con.cursor()
#Creating a cursor object to execute SQL queries on a database table

create_table = "CREATE TABLE IF NOT EXISTS pokemon(Pokemon_No TEXT, Name TEXT, Type_1 TEXT, Type_2 TEXT, Max_CP TEXT, Max_HP TEXT, Image_URL TEXT);"
# creating the table to execute into the database

cur.execute(create_table)

    
con.commit()

con.close

```

```python
# simplified version of ^^^ all that
pogosql = pogo_data.to_sql(create_table, con, if_exists='append', index=False)
```
